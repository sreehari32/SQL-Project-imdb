**Loading IMDB data to dataframe and printing the first 5 rows**

```python
df=pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')  
df.head()
```

```python
#dropping the Poster_Link column from the dataframe
df_clean=df.drop('Poster_Link', axis= 1)
df_clean.head()


#import sqlalchemy and creating sqlite engine

from sqlalchemy import create_engine
engine=create_engine('sqlite://',echo=False)

# exporting the dataframe as a table with name imdb 
df_clean.to_sql("imdb",con=engine)
```

```python
# calculating the number of rows of the table

sql=''' SELECT COUNT(*) 
         FROM imdb'''

df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

```python
# printing top 10 rows of the table
sql=''' SELECT * FROM imdb LIMIT 10'''

df_sql= pd.read_sql_query(sql,con=engine)
df_sql
```

# 1)List the top 10 movies based on their average IMDB rating.

```python
sql ='''
SELECT * 
FROM imdb
ORDER BY IMDB_Rating DESC
LIMIT 10
'''


df1= pd.read_sql_query(sql, con= engine)
df1
```

# Find the number of movies released in each year and order by number of movies released in descending order.

```python
sql='''

select Released_Year, count(*)No_of_movies

FROM imdb
GROUP BY Released_Year
ORDER BY No_of_movies DESC

;
'''


df_sql=pd.read_sql_query(sql,con= engine)
df_sql
```

# Inference
Most number of movies are released in the year 2004 followed by 2009 and 2016

# Retrieve the details of the longest movie in terms of duration.

```python
sql='''

SELECT * 
FROM imdb
WHERE Runtime =(SELECT MAX(Runtime) FROM imdb)
LIMIT 1
;
'''


df_sql= pd.read_sql_query(sql, con=engine)
df_sql
```

# Inference

Crime drama thriller Reservoir Dogs directed by Quentin Tarantino have the longest duration 

# List the top 5 genres with the most movies.

```python
sql= '''
select Genre,count(*)COUNT 
FROM imdb
GROUP BY Genre
ORDER BY COUNT DESC
;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# Inference


The number of movies directed in the Drama genre is double that of the second-highest genre

# Find the actors who have appeared in  most movies.

all the five coloumns where the actor names are given are clubbed and then grouped to see the actor who have appeared in most number of movies


```python
sql= '''
SELECT St, count(*)No_of_films_acted FROM 
( 
SELECT Star1 as St FROM imdb 
UNION ALL
SELECT Star2 FROM imdb
UNION ALL
SELECT Star3 FROM imdb
UNION ALL
SELECT Star4 FROM imdb
) as all_stars

GROUP BY St
ORDER BY No_of_films_acted DESC








;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# Find the actors who have appeared in  most movies as main character.

```python
sql= '''
select Star1,count(*)COUNT 
FROM imdb
GROUP BY Star1
ORDER BY COUNT DESC
;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# Inference

While Tom Hanks has played the main character in the most number of movies, it is Mr. Robert De Niro who has the largest overall number of films in his filmography

# Retrieve the movies that have won the most number of votes

```python
sql= '''
select Series_Title, No_of_Votes
FROM imdb
WHERE No_of_Votes=( SELECT MAX(No_of_Votes) FROM imdb)
;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# List the directors who have directed at least 5 movies.

```python
sql= '''
select Director,count(*)No_of_movies_directed
FROM imdb
GROUP BY Director
HAVING No_of_movies_directed >5
ORDER BY No_of_movies_directed DESC
;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# Find the average rating for each genre.

```python
sql= '''
select Genre, AVG(IMDB_Rating) Avg_rating_of_genre 
FROM imdb
GROUP BY Genre
ORDER BY Avg_rating_of_genre DESC
;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# Retrieve the movies that were released in the 21st century and have a rating above 8.

```python
sql= '''
select * 
FROM imdb
WHERE Released_Year>=2000 AND IMDB_Rating>=8
;


'''


df_sql=pd.read_sql_query(sql,con=engine)
df_sql
```

# List the movies that have both a high rating (above 9) and a large number of votes (above 100,000).
